```python
import pandas as pd

df = pd.read_csv("../data/Superstore.csv")

df.head()
```

```python
import matplotlib.pyplot as plt

region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

print(region_sales)
```

```python
region_sales.plot(kind="bar")

plt.title("Region Wise Sales")
plt.ylabel("Sales")
plt.show()
```

The West region generated the highest sales revenue, contributing significantly to overall business performance. The South region recorded the lowest sales, indicating potential opportunities for market expansion, targeted promotions, or customer acquisition strategies.

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print(category_sales)
```

```python
category_sales.plot(kind="bar")

plt.title("category Wise Sales")
plt.ylabel("Sales")
plt.show()
```
